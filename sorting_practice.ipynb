### Binary Search

```python
# initiate start end
# find mid check if mid value is equal to target then return the mid
# if less change end = mid -1 ; if greater then change start = mid +1
def binary_search(lst, target):
    start = 0
    end = len(lst)-1
    while start <= end:
        mid = (start + end) // 2
        if lst[mid] == target:
            return mid
        elif lst[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return -1
lst = [2, 3, 4, 10, 40, 45, 50, 60, 66]
target = 10
print(binary_search(lst, target))
```

### Bubble Sort

```python
#loop over len of list
# loop angain over the list
# check the adjacent values
# if right value is less than left then swap the value in place
# time complexity O(n2) and space complexity is O(1) -> inplace swapping
def bubble_sort(lst)-> list:
    for passes in range(len(lst)):
        for i in range(len(lst)-1):
            if lst[i+1] < lst[i]:
                lst[i], lst[i+1] = lst[i+1], lst[i]
    return lst
list = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(list)
```

### Selection Sort

```python
# loop over the lst
# take the first value as min and check the rest of data to find min value
# swap the min value to the first
# move the pointer to second value and repeate the process
# time complexity O(n2) and space complexity is O(1)
def selection_sort(lst):
    for i in range(len(lst)-1):
        min_val = lst[i]
        min_pos = i
        for j in range(i+1,len(lst)):
            if lst[j] < min_val:
                min_val = lst[j]
                min_pos = j
        lst[i], lst[min_pos] = lst[min_pos], lst[i]
    return lst
list = [64, 34, 25, 12, 22, 11, 90]
selection_sort(list)
```

### Insertion Sort

```python
# iterate over lst starting with 1 to len(lst)
# point a value (initially 1 pos) iterate over the left side of the list
# check if the pointer is less than a value in left side
# it should be in decending order while checking
# if found the position that can be placed then swap it
# time O(n2) and space O(1)
def insertion_sort(lst):
    for i in range(1, len(lst)):
        current_val = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > current_val:
            lst[j+1] = lst[j]
            j -= 1
        lst[j+1] = current_val
    return lst
list = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(list)
```

### Quick Sort

```python
# Based on divide and conquire method
# divide left and right
# combine left and right by sorting
def merg_lst(left, right):
    i = j = 0
    merged = []
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i+=1
        else:
            merged.append(right[j])
            j+=1
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged
def merg_sort(lst):
    if len(lst) <= 1 :
        return lst
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]
    return merg_lst(merg_sort(left), merg_sort(right))
list = [64, 34, 25, 12, 22, 11, 90]
merg_sort(list)
```

### Quick Sort

```python
# find mid as a pivot variable
# divide as left mid and right based on value less than and greater than pivot value
# make recursive call
def quick_sort(lst):
    if len(lst) <= 1:
        return lst
    pivot = lst[len(lst) // 2]
    left = [val for val in lst if val < pivot]
    mid = [val for val in lst if val == pivot]
    right = [val for val in lst if val > pivot]
    return quick_sort(left) + mid + quick_sort(right)
list = [64, 34, 25, 12, 22, 11, 90]
quick_sort(list)
```
