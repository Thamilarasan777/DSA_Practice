```python
list_names = ['Thamil', 'Arun', 'Deva', 'Gokul','Vijay', 'Ajay']
def collect_names(lst):
    try:   
        output_dict = {}
        for name in lst:
            if name[0] not in output_dict.keys():
                output_dict[name[0]] = [name]
            else:
                output_dict[name[0]] = []
                raise ValueError
    except ValueError:
        ...
    return output_dict
collect_names(list_names)

```

```python
class Box:
    def __init__(self, n):
        self._n = n
    @property
    def value(self):
        return self._n*2
b = Box(5)
print(b.value)
```
