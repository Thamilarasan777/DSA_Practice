Remove extra spaces

```python
import re
text = "GenAI!!! is   AWESOME, isn't it??"
text = text.lower()
print(text)
text = re.sub(r"[^a-z\s]","",text)
print(text)
text = re.sub(r"\s{2,}"," ",text)
print(text)
```

Count Word Frequencies

```python
text = "llm models are large language models"
text = text.lower()
word_count = {}
for word in text.split():
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] +=1
print(word_count)
    
```

Find First Non-Repeating Character

```python
s = "aabbccdde"
current_count = 0
current_char = None
for char in s:
    if char == current_char:
        current_count +=1
    else:
        if current_count == 1:
            print(current_char)
            break
        else:
            current_char = char
            current_count = 1
```

Check if the two string is anagram

```python
s1 = "listen"
s2 = "silent"
print("".join(sorted(s1)) == "".join(sorted(s2)))
```

N- Gram generator

```python
text = "I love gen ai"
text_list = text.split()
n_grams = []
n = 3
for i in range(len(text_list)-(n-1)):
    tup = tuple(text_list[i+j] for j in range(n))
    n_grams.append(tup)
print(n_grams)
```
