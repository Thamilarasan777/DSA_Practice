```python
# create class for a node 
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
n = 3
def add_nodes(n):
    """add nodes and return head"""
    if n == 0:
        return None
    node = Node(n)
    node.next = add_nodes(n-1)
    return node
head = add_nodes(n)
def print_linked_list(head):
    while head != None:
        print(head.value, end="->")
        head = head.next
print_linked_list(head)
```

```python
def insert_head(nodes, head_val):
    newNode = Node(head_val)
    newNode.next = nodes
    return newNode
print_linked_list(head)
print("")
head = insert_head(head, 4)
print_linked_list(head)
```

```python
def insert_tail(head, val):
    if head.next== None:
        newNode = Node(val)
        head.next = newNode
        return head
    return insert_tail(head.next,val)
insert_tail(head, 50)
print_linked_list(head)
```
