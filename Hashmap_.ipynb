```python
# create a class for HashMap
# constructor of len and list[list[(key, val)]] declaration
# define a function to get the hash for a key
# define a function to add elements inside the hashtable
class HashMap:
    def __init__(self,n):
        self.n = n
        self.arr = [[] for i in range(self.n)]
    def get_hash(self, key):
        assci_count = 0
        for char in key:
            assci_count += ord(char)
        return assci_count % self.n
    def __setitem__(self, key, val):
        h_index = self.get_hash(key)
        for index, element in enumerate(self.arr[h_index]):
            if element[0] == key:
                self.arr[h_index][index] = (key, val)
                return
        self.arr[h_index].append((key, val))
    def __getitem__(self, key):
        h_index = self.get_hash(key)
        for element in self.arr[h_index]:
            if len(element) == 2 and element[0] == key:
                return element[1]
    def __delitem__(self, key):
        h_index = self.get_hash(key)
        for index, element in enumerate(self.arr[h_index]):
            if len(element) == 2 and element[0] == key:
                del self.arr[h_index][index]
                return
            
```

```python
hash = HashMap(10)
hash["abc"] = 20
hash["abcd"] = 21
```

```python
del hash["abc"]
hash["abc"]
```
