```python
import re
s = "Hello, World!"
def is_palindrome(s):
    """
    Function to check if the input string is a palindrome.
    
    Parameters:
    s (str): The input string to check.
    
    Returns:
    bool: True if the string is a palindrome, False otherwise.
    """
    # Your code here
    s = re.sub(r"[,\.!\(\)\s]","",s).lower()
    print(s)
    def is_palindrom(txt):
        if len(txt) == 1:
            return True
        if s[0] != s[-1]:
            return False
        return is_palindrom(txt[1:-1])
    return is_palindrom(s)
is_palindrome(s)
```

```python
#Binary search
l = [4,7,9,11,14,16,18,40,45,60]
def binary_search(arr, target):
    start = 0
    end = len(arr)-1
    while start<=end:
        mid = (end + start) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            start = mid + 1
        else:
            end = mid -1
    return -1
binary_search(l, 14)
```

```python
#linear search recurrsion
l = [4,7,9,11,14,16,18,40,45,60]
def recurrsive_linear_search(lst, val, index=0):
    if len(lst) == 0:
        return -1
    if lst[0] == val:
        return index
    return recurrsive_linear_search(lst[1:],val,index+1)
recurrsive_linear_search(l, 60)
```

```python
def selection_sort(lst):
    # Your code goes here
    n = len(lst)
    for passes in range(n):
        min_val = lst[passes]
        index = passes
        for i in range(passes+1,n):
            if lst[i] < min_val:
                index = i
                min_val = lst[i]
        lst[index], lst[passes] = lst[passes], lst[index]
    return lst
selection_sort([64, 25, 12, 22, 11])
```

```python
#count negative numbers in a decenting grid
grid = [[4, 3, 2, 1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]] 
def countNegatives(grid):
    total_negative_count = 0
    column_length = len(grid[0])
    def binary_search(lst,mid_value):
        start = 0
        length = len(lst)
        end = length-1
        for i in range(length):
            mid = (start + end) //2
            if lst[mid]>=mid_value:
                start = mid+1
                continue
            else:
                end = mid
            if mid == start:
                return mid
        return -1
    for item in grid:
        found_index = binary_search(item,0)
        total_negative_count += column_length-found_index if found_index != -1 else 0
    return total_negative_count
        
countNegatives(grid)
```
