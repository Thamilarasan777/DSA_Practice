```python
# factorial
n = 5
def factorial(n):
    if n == 2:
        return n
    return factorial(n-1)*n
print(factorial(n))
```

```python
#Sum of N natural Numbers
def sum_natural_numbers(n):
    if n == 1:
        return n
    return sum_natural_numbers(n-1)+n
print(sum_natural_numbers(5))
```

```python
#print Numbers from 1 to N
def print_numbers_n_to_1(n):
    if n == 0:
        return None
    print(n)
    return print_numbers_n_to_1(n-1)
print_numbers_n_to_1(5)  
```

```python
#print Numbers from 1 to N
def print_numbers_1_to_n(n):
    if n == 0:
        return
    print_numbers_1_to_n(n-1)
    print(n)
     
print_numbers_1_to_n(5)  
```

```python
a = [1,2,3,4,5,6,7]
def reverse_list(list_):
    if len(list_) == 1:
        return list_
    rev = reverse_list(list_[1:])
    rev.append(list_[0])
    return rev
reverse_list(a)
```

```python
# reverse a list
a = [1,2,3,4,5,6,7]
def reverse_list(list_):
    if len(list_) == 1:
        return list_
    rest_reversed = reverse_list(list_[1:])
    rest_reversed.append(list_[0])
    return rest_reversed
reverse_list(a)
```

```python
#find power of a number (X^n)
X = 6
n = 2
def find_power(X,n):
    if n == 1:
        return X
    return find_power(X,n-1)*X
print(find_power(X,n))
```

```python
# check if the string is polindrom or not
txt = "ABCDCBA"
n = len(txt)//2
for i in range(n):
    if txt[i] == txt[len(txt)-1-i]:
        print("ok")
    else:
        print("not ok")
def is_polindrom(text):
    if len(text) == 1:
        return True
    if text[0] != text[-1]:
        return False
    return is_polindrom(text[1:-1])
is_polindrom(txt)
```

```python
#Binary search
l = [4,7,9,11,14,16,18,40,45,60]
def binary_search(arr, target):
    start = 0
    end = len(arr)-1
    while start<=end:
        mid = (end + start) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            start = mid + 1
        else:
            end = mid -1
    return -1
binary_search(l, 14)
```
